## Подгрузка данных

```python
import pandas as pd

data = pd.read_csv('dataset.csv')
```

```python
data.drop(columns=['average_distance', 'country', 'region','bad_dist_count', 'Unnamed: 0', 'city'], inplace=True)
```

```python
data.dropna(inplace=True)
```

```python
data
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
data['max_negative'] = label_encoder.fit_transform(data['max_negative'])
data['max_main'] = label_encoder.fit_transform(data['max_main'])
data['max_positive'] = label_encoder.fit_transform(data['max_positive'])
data['place'] = label_encoder.fit_transform(data['place'])
```

```python
data
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN, MeanShift
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
import hdbscan

scaler = StandardScaler()
X = scaler.fit_transform(data)

cluster_methods = [
    # ("DBSCAN", DBSCAN(eps=0.3, min_samples=5)),
    ("MeanShift", MeanShift(bandwidth=0.9)),
    ("HDBSCAN", hdbscan.HDBSCAN(min_cluster_size=2, min_samples=1))
]

plt.figure(figsize=(15, 5))
for i, (name, method) in enumerate(cluster_methods):
    plt.subplot(1, len(cluster_methods), i + 1)
    if name != "DBSCAN":
        method.fit(X)
        labels = method.labels_
    else:
        labels = method.fit_predict(X)

    silhouette_avg = silhouette_score(X, labels)
    
    plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
    plt.title(f"{name} - Silhouette Score: {silhouette_avg:.2f}")

plt.tight_layout()
plt.show()
```

```python
data['cluster_label'] = hdbscan.HDBSCAN(min_cluster_size=2, min_samples=1).fit_predict(X)
```

```python
data.info()
```

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

y = data['cluster_label'] + 1
X_train, X_test, y_train, y_test = train_test_split(data[['positive_count', 'green_zone',
    'place', 'population', 'negative_count','main_city_count',
    'max_negative','max_main','max_positive']], y, test_size=0.2, random_state=42)

model = xgb.XGBClassifier(
    objective="multi:softmax",  # Многоклассовая классификация
    num_class=len(np.unique(y_train)),  # Количество классов
    max_depth=3,  # Максимальная глубина деревьев
    learning_rate=0.1,  # Скорость обучения
    n_estimators=100  # Количество деревьев
)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
```

```python
import joblib
joblib.dump(model, 'xgboost_model.pkl')
```

```python
loaded_model = joblib.load('xgboost_model.pkl')

y_pred = loaded_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
```
